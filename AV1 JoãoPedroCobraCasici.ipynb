```python
import JoãoPedroCobraCasici as jp
import numpy as np
import matplotlib.pyplot as plt
```

**Questão 1**

```python
def f(x):
    return (4*x + 1)**2 * np.e**(x**2 - 2) - 1

x1 = np.linspace(-1, 1, 200)
y1 = [f(x) for x in x1]

plt.figure(figsize=(6, 4))
plt.plot(x1, y1, label='f(x)')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()


print(jp.secante(f, 0, 0.5))
print(jp.bisseccao(f, -1, 0))
```

**Questão 2**

```python
def g(x):
    return -4*x**2 + 4

x2 = np.linspace(-2, 2, 200)
y2 = [g(x) for x in x2]

a = jp.falsa_posicao(g, -1.5, -0.5)
b = jp.falsa_posicao(g, 0.5, 1.5)

print(a)
print(b)

plt.figure(figsize=(6,4))
plt.plot(x2, y2, label='g(x)')
plt.axhline(0, color = 'red', linestyle='--')
plt.plot(a, 0, 'o', color = 'green')
plt.plot(b, 0, 'o', color = 'green')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True)
plt.show()
```

**Questão 3**

a)

```python
x = np.array([-2.0, 0.0, 1.0, 3.0])
y = np.array([2, 4.5, 5, 5])

b0, b1 = jp.ajuste_linear(x, y)
print(f"Modelo linear: y = {b0:.4f} + {b1:.4f}x")

y_ajuste = b0 + b1 * x

plt.scatter(x, y, color='red', label='Dados')
plt.plot(x, y_ajuste, label='Ajuste Linear', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

b) 

```python
a0, a1, a2 = jp.reg_lin_mult2(x, y)

print(f"Modelo quadrático: y = {a0:.2f} + {a1:.3f}x + {a2:.3f}x²")

x_plot = np.linspace(min(x), max(x), 200)
y_quad = a0 + a1 * x_plot + a2 * x_plot**2

plt.scatter(x, y, color='red', label='Dados')
plt.plot(x_plot, y_quad, label='Ajuste Quadrático', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

c)

```python
y_linear = b0 + b1 * x
res_linear = np.sum((y - y_linear)**2)
tot = np.sum((y - np.mean(y))**2)
r2_linear = 1 - res_linear/tot
print(f"Coeficiente de determinação do modelo linear = {r2_linear:.4f}")

y_quad = a0 + a1 * x + a2 * x**2
res_quad = np.sum((y - y_quad)**2)
r2_quad = 1 - res_quad/tot
print(f"Coeficiente de determinação do modelo quadrático = {r2_quad:.4f}")

# O modelo com maior coeficiente de determinação é o modelo quadrático, pois ele se ajusta melhor aos dados.
# O modelo linear é mais simples, mas o quadrático oferece um ajuste melhor aos dados fornecidos.
```
