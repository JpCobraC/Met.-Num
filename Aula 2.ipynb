```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x): 
    return (0.05 * x**3 - 0.4 * x**2 + 3 * x * np.sin(np.radians(x)))

x = np.linspace(-20, 20, 50)

y = f(x)
```

```python
plt.figure(figsize=(8, 3))
plt.plot(x, y)
plt.title("f(x) = 0.05x³ - 0.4x² + 3x·sin(x)")
plt.grid(True)
plt.show()
```
