```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_regression(x, y):
  n = len(x)
  sum_y = sum(y)
  sum_x = sum(x)
  sum_xy = sum(x*y)
  sum_x2 = sum(x**2)

  b0 = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - sum_x**2)

  return b0, (sum_y - b0*sum_x)/n
```

```python
temp = np.array([3,6,7,8,14,16])
germ = np.array([41,29,27,19,10,11])
```

```python
def draw_points_regression(x, y):
  b0, b1 = linear_regression(x, y)
  x_reg = np.linspace(min(x), max(x), 100)
  y_reg = b0 * x_reg + b1

  plt.scatter(x, y, color='blue', label='Pontos de dados')
  plt.plot(x_reg, y_reg, color='red', label=f'Reta de regressão (y={b0:.4f}x + {b1:.4f})')

  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Regressão Linear')
  plt.legend()
  plt.grid(True)
  return plt
```

```python
linear_regression(temp, germ)
```

```python
x = draw_points_regression(temp, germ)
x.xlabel('x(temp)')
x.ylabel('y(germ)')
x.plot()
```
