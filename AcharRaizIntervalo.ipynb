```python
from numpy import sin, cos
```

```python
def f(x):
  return 5*x**3-11*x**2-14*x+10
```

```python
a = 0
b = 1
tests = 10_000

if f(a) * f(b) < 0:
  for i in range(tests):
    m = (a + b) / 2
    r = f(m)
    if f(a) * r < 0:
      b = m
    else:
      a = m
else:
  m = None

print(m)
```

```python
import matplotlib.pyplot as plt

x_vals = [i/10 for i in range(int(a*10)-5, int(b*10)+5)]
y_vals = [f(x) for x in x_vals]

plt.plot(x_vals, y_vals)

plt.axhline(y=0, color='r', linestyle='--', linewidth=2)

plt.grid(True)
plt.xlabel("x")
plt.ylabel("f(x)", rotation=0)
plt.title("Raiz da função f(x) = 5x³ - 11x² - 14x + 10")

plt.draw()

ymin, ymax = plt.gca().get_ylim()
text_y = ymin + (ymax - ymin) * 0.25

if m is not None:

  plt.plot(m, 0, marker='o', markersize=8, color='green')

  plt.annotate(
      f"Raiz = {m}",
      xy=(m, 0),
      xytext=(m, text_y),
      arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
      fontsize=10,
      ha='center'
  )

plt.show()
```
