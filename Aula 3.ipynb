```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (2 * x**3 + 5 * x**2 - np.sin(np.radians(x)) - 30 )
```

```python
jao = f(1) * f(4)

print(jao)
```

```python
x = np.linspace(1, 4, 4000)

y = f(x)

plt.title("Função")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.plot(x, y)
plt.show()
```

```python
def bissecao(funcao, a, b, tolerancia = 1e-6, max_i=100):
    if funcao(a) * funcao(b) >= 0:
        print("A função deve ter sinais opostos nos limites de iteração")
    iteracao = 0
    while iteracao < max_i and abs(b-a) > tolerancia:
        c = (a+b) / 2
        if f(c) == 0:
            return c, iteracao
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
        iteracao += 1
    return (a+b) / 2, iteracao

print(bissecao(f, 1, 4))
```
